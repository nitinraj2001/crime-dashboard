```python
for row in data_population.values:
    state=row[0]
    base_population=row[1]
    base_litracy=row[3]
    diffPop=row[5]
    diffLitracy=row[6]
    for i in range(2001,2013):
        new_data=np.array([])
        new_data=np.append(new_data,state)
        new_data=np.append(new_data,i)
        change=(i-2001)*diffPop
        curr_population=base_population+change
        new_data=np.append(new_data,curr_population)
        change=(i-2001)*diffLitracy
        curr_litracy=base_litracy+change
        new_data=np.append(new_data,curr_litracy)
        data_crime=data[data['STATE/UT'].str.upper()==state].values
        normalized_crime=(data_crime[i-2001][14]/curr_population)*100000
        new_data=np.append(new_data,normalized_crime)
        new_data=pd.DataFrame(new_data)
        new_data=new_data.transpose()
        new_data=new_data.iloc[0,:].values
        new_data=pd.DataFrame(new_data,index=None, columns=None).transpose()
        new_data.to_csv('children-animation.csv',mode='a', header=False)  
```
